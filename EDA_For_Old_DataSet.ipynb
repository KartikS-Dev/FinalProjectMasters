## Adding The Happiness Scale to the dataset based on rating 
### 0-1 > Very Sad
### 1-2 > sad
### 2-3 > Okay , Neutral Emotion
### 3-4 > Happy,
### 4-4.5 >  Very Happy
### 4.5-5 > Very Happy , Highly Satisfied



```python
tempData=[]
for i in range (0,len(reviewDataSet.index)):
    tempData.append('temp')
reviewDataSet['Happiness_Level'] = tempData

for i in range (0,len(reviewDataSet.index)):
    if reviewDataSet['Rating'][i]<=1.0:
        reviewDataSet['Happiness_Level'][i]='Very Sad'
    elif reviewDataSet['Rating'][i]>1.0 and reviewDataSet['Rating'][i]<=2.0 :
        reviewDataSet['Happiness_Level'][i]='Sad'
    elif reviewDataSet['Rating'][i]>2.0 and reviewDataSet['Rating'][i]<=3.0:
        reviewDataSet['Happiness_Level'][i]='Neutral'
    elif reviewDataSet['Rating'][i]>3.0 and reviewDataSet['Rating'][i]<=4.0:
        reviewDataSet['Happiness_Level'][i]='Happy'
    elif reviewDataSet['Rating'][i]>4.0 and reviewDataSet['Rating'][i]<=4.5:
        reviewDataSet['Happiness_Level'][i]='Very Happy'
    elif reviewDataSet['Rating'][i]>4.5 and reviewDataSet['Rating'][i]<=5.0:
        reviewDataSet['Happiness_Level'][i]='Excellent'
    
reviewDataSet
```
